# Diagrammes Catégoriels - Distribution au sein des Catégories

Jusqu'à présent, nous avons vu comment appliquer une estimation statistique (comme la moyenne ou le nombre d'instances/lignes) à des catégories et les comparer entre elles. Voyons maintenant comment visualiser la distribution au sein des catégories. Nous connaissons déjà displot() qui permet de visualiser la distribution d'une seule feature/variable, nous allons maintenant décomposer cette même distribution par catégorie.

## Importations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Données

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

## Boxplot

Comme décrit dans la vidéo, un diagramme en boîte (ou boîte à moustaches) affiche la distribution à l'aide de l'utilisation de quartiles et d'un écart-interquartile (EI) pour les valeurs aberrantes (outliers).

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df);
```

### Ajout d'un 'hue' pour plus de segmentation

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue='gender')

# Facultatif : positionnement de la légende
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

### Paramètres de stylisation du Boxplot

#### Orientation

```python
# NOTEZ COMMENT NOUS DEVONS ÉCHANGER X ET Y POUR QUE L'ORIENTATION SE CONCRÉTISE !
sns.boxplot(x='math score',y='parental level of education',data=df,orient='h');
```

#### Largeur

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue='gender',width=0.3);
```

## Violinplot

Un diagramme en violon joue un rôle similaire à celui d'une boîte à moustaches. Il montre la distribution des données quantitatives sur plusieurs niveaux d'une (ou plusieurs) variables de catégorie, de sorte que ces distributions puissent être comparées. Contrairement à une boîte à moustaches, dans laquelle toutes les composantes du tracé correspondent à des points de données réels, le diagramme en violon présente une estimation de la densité du noyau de la distribution sous-jacente.

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df);
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,hue='gender');
```

### Paramètres Violinplot

#### split
Lorsque vous utilisez l'imbrication 'hue' avec une feature/variable qui prend deux niveaux, la définition de split sur la valeur True dessinera la moitié d'un violon pour chaque niveau. Il est ainsi plus facile de comparer directement les distributions.

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,hue='gender',split=True);
```

#### inner

Représentation des points de données à l'intérieur du violon. Si inner='box', cela dessine une boîte à moustaches miniature. S'il s'agit de 'quartiles', cela dessine les quartiles de la distribution. Si 'point' ou 'stick', cela affiche chaque point de données sous-jacent. L'utilisation de None dessinera des violons sans rien dedans.

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner=None);
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner='box');
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner='quartile');
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner='stick');
```

#### orientation

```python
# Il suffit de basculer la variable continue sur y et la variable catégorielle sur x
sns.violinplot(x='math score',y='parental level of education',data=df);
```

#### bandwidth

Semblable à l'argument de bande passante (bandwidth) pour le kdeplot

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,bw=0.1);
```

# Diagrammes avancés 

Nous pouvons utiliser un boxenplot et un swarmplot pour obtenir le même effet que le boxplot et le violinplot, mais avec un peu plus d'informations incluses. Soyez prudent lorsque vous utilisez ces tracés, car ils vous obligent souvent à éduquer le spectateur sur la façon dont le diagramme est réellement construit. N'utilisez ces éléments que si vous êtes certain que votre audience comprendra bien la visualisation.

```python
df.head()
```

## swarmplot

```python
sns.swarmplot(x='math score',data=df);
```

```python
sns.swarmplot(x='math score',data=df,size=2);
```

```python
sns.swarmplot(x='math score',y='race/ethnicity',data=df,size=3);
```

```python
sns.swarmplot(x='race/ethnicity',y='math score',data=df,size=3)
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x='race/ethnicity',y='math score',data=df,hue='gender')
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x='race/ethnicity',y='math score',data=df,hue='gender',dodge=True)
```

## boxenplot (letter-value plot)

Papier officiel au sujet de ce diagramme : https://vita.had.co.nz/papers/letter-value-plot.html

Ce style de diagramme a été nommé à l'origine un diagramme de « letter value » car il montre un grand nombre de quantiles définis comme des « valeurs de lettres ». Il s'apparente à une boîte à moustaches pour tracer une représentation non paramétrique d'une distribution dans laquelle toutes les features correspondent à des observations réelles. En traçant plus de quantiles, il fournit plus d'informations sur la forme de la distribution, en particulier dans les queues (tails).

```python
sns.boxenplot(x='math score',y='race/ethnicity',data=df)
```

```python
sns.boxenplot(x='race/ethnicity',y='math score',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='race/ethnicity',y='math score',data=df,hue='gender')
```
