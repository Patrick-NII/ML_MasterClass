# Diagrammes Catégoriels - Estimation statistique au sein des catégories


Nous avons souvent des données **catégorielles**, ce qui signifie que les données se trouvent dans des groupes distincts, tels que des pays ou des entreprises. Il n'y a pas de valeur de pays « entre » les États-Unis et la France et il n'y a pas de valeur d'entreprise « entre » Google et Apple, contrairement aux données continues où nous savons que des valeurs peuvent exister entre des points de données, tels que l'âge ou le prix.

Pour commencer par les diagrammes catégoriels, nous allons nous concentrer sur l'estimation statistique au sein des catégories. En gros, cela signifie que nous allons rapporter visuellement certaines statistiques (telles que la moyenne ou le nombre d'instances/lignes) dans un graphique. Nous savons déjà comment obtenir ces données avec Pandas, mais il est souvent plus facile de comprendre les données si nous les traçons.

## Importations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Données

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

# Diagramme de comptage - Countplot()

Un diagramme simple, il montre simplement le nombre total de lignes par catégorie. 

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='division',data=df);
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df);
```

### Répartition/Séparation au sein d'une autre catégorie avec 'hue'

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df,hue='training level');
```

**NOTE : Vous pouvez toujours modifier la palette à votre goût dans n'importe quelle [colormap](https://matplotlib.org/stable/gallery/color/colormap_reference.html) Matplotlib**

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df,hue='training level',palette='Set1');
```

```python
plt.figure(figsize=(10,4),dpi=200)
# 'Paired' serait un bon choix s'il y avait un saut distinct de 0 et 1 à 2 et 3.
sns.countplot(x='level of education',data=df,hue='training level',palette='Paired');
```

# Diagramme en barres - barplot()

Jusqu'à présent, nous avons vu l'axe Y par défaut pour un comptage (similaire à un appel .groupby (x_axis) .count () avec Pandas). Nous pouvons développer nos visualisations en spécifiant une fonction continue spécifique pour l'axe Y. Gardez à l'esprit que vous devez être prudent avec ces diagrammes, car ils peuvent impliquer une continuité de relation le long de l'axe Y là où il n'y en a pas.

```python
plt.figure(figsize=(10,6),dpi=200)
# Par défaut barplot() montrera la moyenne
# Informations sur le trait noir : https://stackoverflow.com/questions/58362473/what-does-black-lines-on-a-seaborn-barplot-mean
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd');
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division');
```

```python
plt.figure(figsize=(12,6),dpi=100)

# https://stackoverflow.com/questions/30490740/move-legend-outside-figure-in-seaborn-tsplot
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division')

plt.legend(bbox_to_anchor=(1.05, 1));
```
