# Diagrammes de Dispersion

Les diagrammes de dispersion ou nuages de points (scatter plots) permettent de montrer comment différentes variables ou features sont liées les unes aux autres. Le thème principal de tous les types de diagrammes relationnels, c'est qu'ils montrent la façon dont les variables/features sont interconnectées les unes aux autres. Il existe de nombreux types de diagrammes qui peuvent être utilisés pour cela. Nous allons donc explorer scatterplot() ainsi que les paramètres généraux de Seaborn applicables à d'autres types de tracés.

## Données

Nous allons utiliser certaines données générées à partir de : http://roycekimmons.com/tools/generated_data



```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
df.info()
```

# Scatterplot

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

## Connexion à l'objet Figure dans Matplotlib

**Notez que Matplotlib est toujours connecté à Seaborn en dessous (même sans importer mMtplotlib.pyplot), puisque Seaborn lui-même fait directement un appel Figure avec Matplotlib. Nous pouvons importer matplotlib.pyplot et effectuer des appels pour affecter directement la figure Seaborn.**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df)
```

## Paramètres Seaborn

Les paramètres hue (de teinte) et palette sont généralement disponibles pour de nombreux appels de tracés dans Seaborn.

### hue

Les points de couleur sont basés sur une feature/caractéristique de catégorie dans le DataFrame.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
```

### Palette

**Choix d'une palette avec cmap (cartes de couleur) de Matplotlib: https://matplotlib.org/stable/tutorials/colors/colormaps.html**

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='viridis')
```

## Paramètres Scatterplot

Ces paramètres sont plus spécifiques à l'appel scatterplot()

### size

Permet de dimensionner en fonction d'une autre colonne.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

### markers

**Utilisez *s=* si vous voulez que la taille du marqueur soit une valeur entière uniforme.**

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)
```

### style

Choisissez automatiquement des styles basés sur une autre feature/variable catégorielle dans l'ensemble de données. Utilisez éventuellement le paramètre **markers=** pour passer une liste de choix de marqueurs basée sur Matplotlib, par exemple : ['*','+','o']

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education')
```

```python
plt.figure(figsize=(12,8))
# Parfois, c'est bien de faire les deux, la teinte (hue) et le style, sur la même colonne.
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)
```

## Exporter une Figure Seaborn

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)

# Appeler savefig dans la même cellule
plt.savefig('example_scatter.jpg')
```
