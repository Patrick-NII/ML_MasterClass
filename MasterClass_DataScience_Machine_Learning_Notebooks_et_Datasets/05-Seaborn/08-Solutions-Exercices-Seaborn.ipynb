### Mission 1 : Reproduire le diagramme de dispersion (nuage de points) illustré ci-dessous

**Le diagramme de dispersion tente de montrer la relation entre les jours travaillés et l'âge de la personne (DAYS_BIRTH) pour les personnes qui n'étaient pas sans emploi. Notez que pour reproduire ce graphique, vous devez d'abord supprimer les personnes sans emploi du dataset. Notez également le signe de l'axe, ils sont tous deux transformés pour être positifs. Enfin, n'hésitez pas à ajuster les paramètres *alpha* et *linewidth* dans le nuage de points, car il y a de nombreux points empilés les uns sur les autres.** 

<img src="https://moncoachdata.com/wp-content/uploads/MonCoachData-cours/Solutions Seaborn/scatterplot_solution.jpg">


```python
import warnings
```

```python
warnings.simplefilter('ignore')
```

```python
plt.figure(figsize=(12,8))

# RETIRER LES PERSONNES SANS EMPLOI
employed = df[df['DAYS_EMPLOYED']<0]

# RENDRE LES 2 COLONNES POSITIVES
employed['DAYS_EMPLOYED'] = -1*employed['DAYS_EMPLOYED']
employed['DAYS_BIRTH'] = -1*employed['DAYS_BIRTH']

# Avec autant de points, alpha est minuscule, cela peut indiquer qu'un
# nuage de points n'est peut-être pas le meilleur choix !

sns.scatterplot(y='DAYS_EMPLOYED',x='DAYS_BIRTH',data=employed,
                alpha=0.01,linewidth=0)

plt.savefig('scatterplot_solution.jpg')
```

### Mission 2 : Reproduire le diagramme de distribution illustré ci-dessous :

<img src="https://moncoachdata.com/wp-content/uploads/MonCoachData-cours/Solutions Seaborn/histplot_solution.png">

**Notez que vous devrez déterminer comment calculer "Âge en années" à partir de l'une des colonnes du dataframe. Réfléchissez bien à cela. Ne vous inquiétez pas trop si vous ne parvenez pas à reproduire exactement le style.**

```python
plt.figure(figsize=(8,4))

df['YEARS'] = -1*df['DAYS_BIRTH']/365
sns.histplot(data=df,x='YEARS',linewidth=2,edgecolor='black',
             color='red',bins=45,alpha=0.4)
plt.xlabel("Âge en années")
plt.ylabel("Nombre")

plt.savefig('histplot_solution.png')
```

### Mission 3 : Reproduire le diagramme catégoriel illustré ci-dessous :

<img src='https://moncoachdata.com/wp-content/uploads/MonCoachData-cours/Solutions Seaborn/boxplot_solution.png'>

**Ce diagramme présente des informations uniquement pour la *moitié inférieure* des salariés de cet ensemble de données. Il affiche les boîtes à moustaches de chaque catégorie de la colonne 'NAME_FAMILY_STATUS' (situation familiale) pour afficher la distribution de leur revenu total. Le "hue" est la colonne 'FLAG_OWN_REALTY' (propriétaire ou non - Y/N). Remarque : Vous devrez ajuster ou ne prendre qu'une partie du dataframe *avant* de reproduire ce tracé.**

```python
plt.figure(figsize=(12,5))

bottom_half_income = df.nsmallest(n=int(0.5*len(df)),columns='AMT_INCOME_TOTAL')
sns.boxplot(x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',data=bottom_half_income,hue='FLAG_OWN_REALTY')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,title='FLAG_OWN_REALTY')
plt.title('Revenus totaux par situation familiale pour la moitié inférieure des salariés')

plt.savefig('boxplot_solution.png')
```
