# Traitement des Données Catégorielles

De nombreux modèles de Machine Learning ne peuvent pas traiter les ensembles de données catégorielles comme des chaînes de caractères String. Par exemple, la régression linéaire ne peut pas appliquer un coefficient bêta à des couleurs comme "rouge" ou "bleu". Au lieu de cela, nous devons convertir ces catégories en variables "dummy" (factices), autrement connu sous le nom d'encodage "one-hot".

## Colonne numérique à catégorielle

Nous devons être prudents lorsqu'il s'agit d'encoder des catégories sous forme de nombres. Nous voulons nous assurer que la relation numérique a un sens pour un modèle. Par exemple, l'encodage de "MSSubClass" est essentiellement un code numérique par classe :

    MSSubClass: Identifies the type of dwelling involved in the sale.	

        20	1-STORY 1946 & NEWER ALL STYLES
        30	1-STORY 1945 & OLDER
        40	1-STORY W/FINISHED ATTIC ALL AGES
        45	1-1/2 STORY - UNFINISHED ALL AGES
        50	1-1/2 STORY FINISHED ALL AGES
        60	2-STORY 1946 & NEWER
        70	2-STORY 1945 & OLDER
        75	2-1/2 STORY ALL AGES
        80	SPLIT OR MULTI-LEVEL
        85	SPLIT FOYER
        90	DUPLEX - ALL STYLES AND AGES
       120	1-STORY PUD (Planned Unit Development) - 1946 & NEWER
       150	1-1/2 STORY PUD - ALL AGES
       160	2-STORY PUD - 1946 & NEWER
       180	PUD - MULTILEVEL - INCL SPLIT LEV/FOYER
       190	2 FAMILY CONVERSION - ALL STYLES AND AGES

Le nombre ne semble pas avoir de relation avec les autres nombres. Si 30 > 20 est vrai, il n'est pas vraiment logique que "1-STORY 1945 & OLDER" > "1-STORY 1946 & NEWER ALL STYLES". Gardez à l'esprit que ce n'est pas toujours le cas, par exemple les sièges de 1ère classe par rapport aux sièges de 2ème classe encodés avec les chiffres 1 et 2. Assurez-vous de bien comprendre votre ensemble de données pour examiner ce qui doit être converti/changé.

## Création de variables Dummy à partir des objets Colonne

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.select_dtypes.html

```python
df.select_dtypes(include='object')
```

```python
df_nums = df.select_dtypes(exclude='object')
df_objs = df.select_dtypes(include='object')
```

```python
df_nums.info()
```

```python
df_objs.info()
```

### Conversion

```python
df_objs = pd.get_dummies(df_objs,drop_first=True)
```

```python
final_df = pd.concat([df_nums,df_objs],axis=1)
```

```python
final_df
```
