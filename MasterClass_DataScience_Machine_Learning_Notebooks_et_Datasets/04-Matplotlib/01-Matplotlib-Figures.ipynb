## Création d'une Figure

L'idée principale de l'utilisation de la méthode orientée objet plus formelle est de créer des objets de figure, puis d'appeler simplement des méthodes ou des attributs à partir de cet objet. Cette approche est plus agréable lorsqu'il s'agit d'une toile comportant plusieurs tracés.

```python
# Crée un canvas vierge
fig = plt.figure()
```

**REMARQUE : TOUTES LES COMMANDES DOIVENT ALLER DANS LA MÊME CELLULE !**

Pour commencer, nous créons une instance de Figure. Ensuite, nous pouvons ajouter des axes à cette figure :

```python
# Créer une Figure (canevas vide)
fig = plt.figure()

# Ajouter un ensemble d'axes à la figure
axes = fig.add_axes([0, 0, 1, 1]) # gauche, bas, largeur, hauteur (plage de 0 à 1)

# Tracer sur cet ensemble d'axes
axes.plot(x, y)

plt.show()
```

```python
# Créer une Figure (canevas vide)
fig = plt.figure()

# Ajouter un ensemble d'axes à la figure
axes = fig.add_axes([0, 0, 1, 1]) # gauche, bas, largeur, hauteur (plage de 0 à 1)

# Tracer sur cet ensemble d'axes
axes.plot(a, b)

plt.show()
```

### Ajout d'un autre ensemble d'axes à la figure

Jusqu'à présent, nous n'avons vu qu'un seul ensemble d'axes sur cet objet Figure, mais nous pouvons continuer à y ajouter de nouveaux axes à n'importe quel emplacement et taille que nous voulons. Nous pouvons ensuite tracer ce nouvel ensemble d'axes.

```python
type(fig)
```

Le code est un peu plus compliqué, mais l'avantage c'est que nous avons maintenant le contrôle total de l'emplacement des axes du tracé et que nous pouvons facilement ajouter plus d'un axe à la figure. Notez comment nous tracons a, b deux fois ici.

```python
# Créer un canvas vide
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Grande figure
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5]) # Plus petite figure

# Figure plus grande Axes 1
axes1.plot(a, b)

# Utiliser set_ pour ajouter des éléments aux axes de la figure
axes1.set_xlabel('Label X')
axes1.set_ylabel('Label Y')
axes1.set_title('Grande Figure')

# Insérer Figure Axes 2
axes2.plot(a,b)
axes2.set_title('Petite Figure');
```

Déplacons la petite figure et modifions ses paramètres.

```python
# Créer un canvas vide
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Grande figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Plus petite figure

# Figure plius grande Axes 1
axes1.plot(a, b)

# Utiliser set_ pour ajouter des éléments aux axes de la figure
axes1.set_xlabel('Label X')
axes1.set_ylabel('Label Y')
axes1.set_title('Grande Figure')

# Insérer Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom Avant');
```

Vous pouvez ajouter autant d'axes sur la même figure que vous le souhaitez, même en dehors de la figure principale si la longueur et la largeur correspondent à cela.

```python
# Crée un canvas vide
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Figure complète
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Plus petite figure
axes3 = fig.add_axes([1, 1, 0.25, 0.25]) # Commence dans le coin supérieur droit !

# Figure plus grande Axes 1
axes1.plot(a, b)

# Utiliser set_ pour ajouter des éléments aux axes de la figure
axes1.set_xlabel('Label X')
axes1.set_ylabel('Label Y')
axes1.set_title('Grande Figure')

# Insérert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom Avant');

# Insérer Figure Axes 3
axes3.plot(a,b)
```

## Exportation d'une Figure

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a,b)
axes1.set_xlabel('X')

# bbox_inches = 'tight' s'assure automatiquement que le cadre de délimitation est correct
fig.savefig('figure.png', bbox_inches = 'tight')
```
