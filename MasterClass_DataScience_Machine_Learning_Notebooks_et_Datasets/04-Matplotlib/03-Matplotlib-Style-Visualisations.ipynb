# Style avec Matplotlib

Importez le module `matplotlib.pyplot` sous le nom de `plt` (de la manière la plus simple) :

```python
# RREUR COURANTE !
# N'OUBLIEZ PAS la partie .PYPLOT

import matplotlib.pyplot as plt
```

## Les Données

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = 2 * x
```

## Légendes

Vous pouvez utiliser l'argument mot-clé **label="label text"** lorsque des tracés ou d'autres objets sont ajoutés à la figure, puis utiliser la méthode **legend** sans arguments pour ajouter la légende à la figure :

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

Remarquez comment la légende pourrait potentiellement chevaucher une partie du graphique réel !

La fonction **legend** prend un argument mot-clé optionnel **loc** qui peut être utilisé pour spécifier où la légende doit être dessinée dans la figure. Les valeurs autorisées de **loc** sont des codes numériques pour les différents endroits où la légende peut être dessinée. Voir la [page de documentation](http://matplotlib.org/users/legend_guide.html#legend-location) pour plus de détails. Certaines des valeurs **loc** les plus courantes sont :

```python
# Beaucoup d'options...

ax.legend(loc=1) # coin supérieur droit
ax.legend(loc=2) # coin supérieur gauche
ax.legend(loc=3) # coin inférieur gauche
ax.legend(loc=4) # coin inférieur droit

# ... de nombreuses autres options sont disponibles

# Les plus courants à choisir
ax.legend(loc=0) # laisser matplotlib décider de l'emplacement optimal
fig
```

```python
ax.legend(loc=(1.1,0.5)) # réglage manuel de l'emplacement
fig
```

## Définition des couleurs, des largeurs de ligne, des types de ligne

Matplotlib vous offre beaucoup d'options de personnalisation des couleurs, des largeurs de ligne et des types de lignes.

Il y a la syntaxe de base comme MATLAB (que je vous suggère d'éviter d'utiliser à moins que vous ne vous sentiez déjà vraiment à l'aise avec MATLAB). Concentrez-vous plutôt sur les paramètres des mots-clés.


### Aperçu rapide

### Couleurs avec une syntaxe comme MatLab

Avec matplotlib, nous pouvons définir les couleurs des lignes et autres éléments graphiques de plusieurs façons. Tout d'abord, nous pouvons utiliser la syntaxe de type MATLAB, où `"b"` signifie bleu (blue), `"g"` signifie vert (green), etc. L'API de MATLAB pour la sélection des styles de lignes est également prise en charge : où, par exemple, `"b.-"` signifie une ligne bleue avec des points :

```python
# MATLAB couleur et style de la ligne de style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # ligne bleue avec des points
ax.plot(x, x**3, 'g--') # ligne verte en pointillés
```

## Approche suggérée : utiliser des arguments par mot-clé



### Couleurs avec le paramètre color

Nous pouvons également définir les couleurs par leur nom ou leurs codes hexadécimaux RGB et éventuellement fournir une valeur alpha à l'aide des arguments mots-clés `color` et `alpha`. Alpha indique une opacité.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # à moitié transparent
ax.plot(x, x+2, color="#8B008B")        # Code hexadécimal RGB
ax.plot(x, x+3, color="#FF8C00")        # Code hexadécimal RGB
```

### Styles de lignes et de marqueurs

#### Largeur de trait

Pour modifier la largeur de la ligne, nous pouvons utiliser l'argument mot-clé `linewidth` ou `lw`.

```python
fig, ax = plt.subplots(figsize=(12,6))

# Utiliser linewidth ou lw
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10)
```

#### Styles de ligne

Il y a de nombreux styles de lignes à choisir, voici la sélection :

```python
# options de type de ligne possibles ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # trait
ax.plot(x, x-2, color="green", lw=3, ls='-.') # tiret et point
ax.plot(x, x-3, color="green", lw=3, ls=':') # points
ax.plot(x, x-4, color="green", lw=3, ls='--') # tirets
```

#### Tiret de ligne personnalisé

La séquence de tirets est une séquence de floats de longueur égale décrivant la longueur des tirets et des espaces en points.

Par exemple, (5, 2, 1, 2) décrit une séquence de tirets de 5 points et 1 point séparés par deux espaces de points.

Tout d'abord, nous voyons que nous pouvons réellement « saisir » la ligne de la commande .plot()

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x,x)

print(type(lines))
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# tirets personnalisés
lines = ax.plot(x, x+8, color="black", lw=5)
lines[0].set_dashes([10, 10]) # format : longueur de la ligne, longueur de l'espace
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# tirets personnalisés
lines = ax.plot(x, x+8, color="black", lw=5)
lines[0].set_dashes([1, 1,1,1,10,10]) # format : longueur de la ligne, longueur de l'espace
```

### Marqueurs

Nous avons techniquement toujours tracé des points, et matplotlib a tracé automatiquement une ligne entre ces points pour nous. Examinons comment placer des marqueurs à chacun de ces points.



#### Style des marqueurs


Une vaste liste de types de marqueurs peut être trouvée ici : https://matplotlib.org/3.2.2/api/markers_api.html

```python
fig, ax = plt.subplots(figsize=(12,6))

# Utiliser marker avec un code
# Utiliser markersize ou ms pour la taille

ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) # ms peut être utilisé pour markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # mettre linewidth à 0 pour voir seulement les marqueurs
ax.plot(x, x-4,marker='1',ms=20)
```

#### Bords, épaisseur, taille et style des marqueurs personnalisés

```python
fig, ax = plt.subplots(figsize=(12,6))

# taille et couleur du marqueur
ax.plot(x, x, color="black", lw=1, ls='-', marker='s', markersize=20, 
        markerfacecolor="red", markeredgewidth=8, markeredgecolor="blue");
```
