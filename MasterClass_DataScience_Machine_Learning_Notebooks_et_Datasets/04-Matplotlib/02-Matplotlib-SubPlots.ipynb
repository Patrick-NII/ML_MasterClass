## plt.subplots()

**REMARQUE : Assurez-vous de placer toutes les commandes dans la même cellule que dans ce notebook et cette vidéo !**

L'objet plt.subplots() agira comme un gestionnaire d'axes plus automatique. Il est ainsi beaucoup plus facile d'afficher plusieurs parcelles côte à côte.


Notez comment nous utilisons la décomposition de tuple pour récupérer à la fois l'objet Figure et un tableau NumPy d'axes :

```python
# Utilisation similaire à plt.figure() sauf que l'on utilise la décomposition de tuple pour récupérer la Figure et les axes
fig, axes = plt.subplots()

# Maintenant, utilisez l'objet axes pour ajouter des éléments au tracé
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titre'); #; Masque Out[]
```

## Ajout de lignes et de colonnes


Vous pouvez alors spécifier le nombre de lignes et de colonnes lors de la création de l'objet subplots() :

```python
# Canvas vide de sous-graphes 1 par 2
fig, axes = plt.subplots(nrows=1, ncols=2)
```

## Tracé sur les objets Axes


Comme auparavant, nous utilisons .plot() sur les objets Axes, et nous pouvons également utiliser les méthodes .set_ sur chaque axe.


Explorons ceci, assurez-vous que tout cela se trouve dans la même cellule :

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for axe in axes:
    axe.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
# NOTE! Ceci retourne un tableau à 2 dimensions
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

Un problème courant avec matplolib est le chevauchement de sous-graphiques ou de figures. Nous pouvons utiliser la méthode **fig.tight_layout()** ou **plt.tight_layout()**, qui ajuste automatiquement les positions des axes sur le canevas de la figure afin qu'il n'y ait pas de chevauchement du contenu :

```python
# NOTE! Cela renvoie un tableau à 2 dimensions
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)  

plt.tight_layout()
```

## Paramètres sur subplots ()

Rappelons que nous avons à la fois l'objet Figure et les axes. Cela signifie que nous pouvons modifier les propriétés sur deux niveaux.

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# PARAMÉTREZ D'ABORD VOS AXES

# Paramètres au niveau de Axes
axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Titre')


axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Titre')
axes[1][1].set_xlabel('1 1 Label X')

axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# PUIS DÉFINIR LES PARAMÈTRES DE LA FIGURE GLOBALE

# Paramètres au niveau de la Figure
fig.suptitle("Niveau Figure",fontsize=16)

plt.show()
```

### Espacement manuel sur subplots()


Utilisez .subplots_adjust pour ajuster l'espacement manuellement.


Tous les détails ici : https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

Exemple tiré du lien :

* left = 0.125  # le côté gauche des sous-graphiques de la figure
* right = 0.9   # le côté droit des sous-graphiques de la figure
* bottom = 0.1  # le bas des sous-graphiques de la figure
* top = 0.9     # le haut des sous-graphiques de la figure
* wspace = 0.2  # la quantité de largeur réservée à l'espace entre les sous-graphiques, exprimée en fraction de la largeur moyenne de l'axe
* hspace = 0.2  # la quantité de hauteur réservée à l'espace entre les sous-graphiques, exprimée en fraction de la hauteur moyenne de l'axe



```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Paramètres au niveau de Axes
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# Utiliser left,right,top, bottom pour étirer les sous-graphiques
# Utiliser wspace,hspace pour ajouter un espacement entre les sous-graphiques
fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.1,)

plt.show()
```

## Exportation de plt.subplots()

```python
# NOTE ! Cela renvoie un tableau à 2 dimensions
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig('subplots.png',bbox_inches='tight')

plt.show()
```
