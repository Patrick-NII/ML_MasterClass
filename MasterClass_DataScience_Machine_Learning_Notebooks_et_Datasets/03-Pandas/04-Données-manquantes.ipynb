## A quoi ressemblent les objets Null/NA/nan :

Source: https://github.com/pandas-dev/pandas/issues/28095

Une nouvelle valeur pd.NA (singleton) est introduite pour représenter les valeurs manquantes scalaires. Jusqu'à présent, Pandas utilisait plusieurs valeurs pour représenter les données manquantes : np.nan est utilisé à cet effet pour les données de type float, np.nan ou None pour les données de type object et pd.NaT pour les données de type date. L'objectif de pd.NA est de fournir un indicateur de données "manquantes" qui peut être utilisé de manière cohérente pour tous les types de données. pd.NA est actuellement utilisé par les types de données booléennes et entières annulables, ainsi que par le nouveau type de données string.

```python
import numpy as np
import pandas as pd
```

```python
np.nan
```

```python
pd.NA
```

```python
pd.NaT
```

## Remplissage avec interpolation

Soyez prudent avec cette technique, vous devez essayer de vraiment comprendre si ce choix est valable ou non pour vos données. Vous devez également noter qu'il existe plusieurs méthodes disponibles, la méthode par défaut étant une méthode linéaire.

Documentation complète sur cette méthode :
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html

```python
airline_tix = {'first':100,'business':np.nan,'economy-plus':50,'economy':30}
```

```python
ser = pd.Series(airline_tix)
```

```python
ser
```

```python
ser.interpolate()
```

```python
ser.interpolate(method='linear')
```

```python
df = pd.DataFrame(ser,columns=['Price'])
```

```python
df
```

```python
df.interpolate()
```

```python
df = df.reset_index()
```

```python
df
```

```python
df.interpolate(method='spline',order=2)
```
