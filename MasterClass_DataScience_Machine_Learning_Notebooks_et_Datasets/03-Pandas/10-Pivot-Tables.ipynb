## Méthode pivot()

La méthode du pivot remodèle les données en fonction des valeurs des colonnes et de la réaffectation de l'index. Gardez à l'esprit qu'il n'est pas toujours utile de faire pivoter les données. Dans nos leçons de Machine Learning, nous verrons que nos données n'ont pas besoin d'être pivotées. Les méthodes de pivotement sont principalement destinées à l'analyse, la visualisation et l'exploration de données.

**Quel type de question un pivot aide-t-il à répondre ?**

Imaginez que nous voulions savoir combien de licences de chaque type de produit Google a acheté ? Actuellement, la façon dont les données sont formatées est difficile à lire. Pivotons-les pour que ce soit plus clair, nous prendrons un sous-ensemble des données pour la question en cours.

```python
# Prenons un sous-ensemble, sinon nous obtiendrons une erreur due à des lignes et des données en double
licenses = df[['Company','Product','Licenses']]
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

## La méthode pivot_table()

Comme la méthode pivot(), la méthode pivot_table() permet d'ajouter des fonctions d'agrégation à un appel pivot.

```python
df
```

```python
# Notez que sommer la feature Account Number n'a aucun sens
pd.pivot_table(df,index="Company",aggfunc='sum')
```

```python
# Soit vous saisissez les colonnes
pd.pivot_table(df,index="Company",aggfunc='sum')[['Licenses','Sale Price']]
```

```python
# Ou les énoncer comme des valeurs souhaitées avec le paramètre values
pd.pivot_table(df,index="Company",aggfunc='sum',values=['Licenses','Sale Price'])
```

```python
df.groupby('Company').sum()[['Licenses','Sale Price']]
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=['Sale Price'],aggfunc='sum')
```

Les colonnes sont facultatives - elles fournissent un moyen supplémentaire de segmenter les valeurs réelles qui vous intéressent. Les fonctions d'agrégation sont appliquées aux valeurs que vous listez.

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc=[np.sum])
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc=[np.sum],fill_value=0)
```

```python
# Possibilité d'ajouter plusieurs fonctions agg
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],
               aggfunc=[np.sum,np.mean],fill_value=0)
```

```python
# Possibilité d'ajouter plusieurs colonnes
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price","Licenses"],columns=["Product"],
               aggfunc=[np.sum],fill_value=0)
```

```python
# Possibilité d'ajouter plusieurs colonnes
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price","Licenses"],
               aggfunc=[np.sum],fill_value=0)
```

```python
# obtenir finalement "TOUT" avec margins = True
# Possibilité d'ajouter plusieurs colonnes
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price","Licenses"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
pd.pivot_table(df,index=["Account Manager","Status"],values=["Sale Price"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```
