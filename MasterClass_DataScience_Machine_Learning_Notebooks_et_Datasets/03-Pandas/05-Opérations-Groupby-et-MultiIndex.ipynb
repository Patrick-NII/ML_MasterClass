# Opérations Groupby et Index Multi-level

```python
import numpy as np
import pandas as pd
```

## Données

```python
df = pd.read_csv('mpg.csv')
```

```python
df
```

## Méthode groupby()

```python
# Crée un objet groupby en attente d'une méthode d'agrégation
df.groupby('model_year')
```

### Ajout d'un appel à une méthode d'agrégation

Pour utiliser un objet groupé, vous devez indiquer à Pandas comment vous souhaitez agréger les données.

Options courantes :

    mean(): Calcule la moyenne des groupes
    sum(): Calcule la somme des valeurs du groupe
    size(): Calcule les tailles de groupe
    count(): Calcule le nombre du groupe
    std(): Écart-type des groupes
    var(): Calcule la variance des groupes
    sem(): Erreur standard de la moyenne des groupes
    describe(): Génère des statistiques descriptives
    first(): Calcule la première des valeurs du groupe
    last(): Calcule la dernière des valeurs du groupe
    nth(): Prend la n-ième valeur, ou un sous-ensemble si n est une liste
    min(): Calcule le minimum des valeurs du groupe
    max(): Calcule le maximum des valeurs du groupe
    
Liste complète dans la documentation en ligne : https://pandas.pydata.org/docs/reference/groupby.html

```python
# model_year devient l'index ! Ce n'est PAS un nom de colonne, c'est maintenant le nom de l'index.
df.groupby('model_year').mean()
```

## Saisir des éléments à partir de la section transversale avec .xs()

Cette méthode prend un argument `key` pour sélectionner les données d'un niveau
particulier d'un MultiIndex.

Paramètres
----------
    key : label or tuple of label
        Étiquette contenue dans l'index, ou partiellement dans un MultiIndex.
    axis : {0 or 'index', 1 or 'columns'}, default 0
        Axe sur lequel récupérer la section transversale.
    level : object, defaults to first n levels (n=1 or len(key))
        Dans le cas d'une clé partiellement contenue dans un MultiIndex, indiquer
        quels niveaux sont utilisés. Les niveaux peuvent être référencés par le label ou la position.

```python
year_cyl.xs(key=70,axis=0,level='model_year')
```

```python
# Valeurs moyennes des colonnes pour 4 cylindres par an
year_cyl.xs(key=4,axis=0,level='cylinders')
```

**Note importante !**

Gardez à l'esprit qu'il est généralement plus facile de filtrer les valeurs **avant** d'exécuter un appel groupby(), vous devez donc essayer de filtrer toutes les valeurs/catégories que vous ne voulez pas utiliser. Par exemple, il est beaucoup plus facile de supprimer les voitures **4** cylindres avant l'appel groupby(), très difficile de faire ce genre de chose après un group by.

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

# Avancé : méthode agg()

La méthode agg() vous permet de personnaliser les fonctions d'agrégation que vous souhaitez pour chaque catégorie.

```python
df
```

### agg() sur un DataFrame

```python
# Ces chaînes doivent correspondre aux noms des méthodes intégrées
df.agg(['median','mean'])
```

```python
df.agg(['sum','mean'])[['mpg','weight']]
```

### Spécifier les méthodes d'agrégation par colonne

**agg()** est très puissant, car il vous permet de passer un dictionnaire dont les clés sont les colonnes et les valeurs sont une liste de méthodes d'agrégation.

```python
df.agg({'mpg':['median','mean'],'weight':['mean','std']})
```

### agg() avec groupby()

```python
df.groupby('model_year').agg({'mpg':['median','mean'],'weight':['mean','std']})
```
