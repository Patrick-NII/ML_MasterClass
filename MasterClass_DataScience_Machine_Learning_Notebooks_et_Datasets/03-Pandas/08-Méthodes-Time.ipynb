# Méthodes Time

## Revue de Python Datetime

Le langage Python de base, en dehors de Pandas, contient une bibliothèque de datetime :

```python
from datetime import datetime
```

```python
# Pour illustrer l'ordre des arguments
my_year = 2017
my_month = 1
my_day = 2
my_hour = 13
my_minute = 30
my_second = 15
```

```python
# 2 janvier 2017
my_date = datetime(my_year,my_month,my_day)
```

```python
# Par défaut 0:00
my_date 
```

```python
# 2 janvier 2017 à 13:30:15
my_date_time = datetime(my_year,my_month,my_day,my_hour,my_minute,my_second)
```

```python
my_date_time
```

Vous pouvez saisir n'importe quelle partie de l'objet datetime que vous voulez

```python
my_date.day
```

```python
my_date_time.hour
```

## Formatage personnalisé des chaînes Time

Parfois, les dates peuvent avoir un format non standard, heureusement vous pouvez toujours spécifier à Pandas le format. Vous devez également noter que cela peut accélérer la conversion, donc cela peut valoir la peine de le faire même si Pandas est capable d'analyser tout seul.

Un tableau complet des codes est disponible ici : https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes

```python
style_date = '12--Dec--2000'
```

```python
pd.to_datetime(style_date, format='%d--%b--%Y')
```

```python
strange_date = '12 Dec 2000'
```

```python
pd.to_datetime(strange_date)
```

## Tentative d'analyse automatique des dates

**parse_dates** - bool ou liste d'int ou noms ou liste de listes ou dict, par défaut False
Le comportement est le suivant :

    boolean. If True -> try parsing the index.

    list of int or names. e.g. If [1, 2, 3] -> try parsing columns 1, 2, 3 each as a separate date column.

    list of lists. e.g. If [[1, 3]] -> combine columns 1 and 3 and parse as a single date column.

    dict, e.g. {‘foo’ : [1, 3]} -> parse columns 1, 3 as date and call result ‘foo’

    If a column or index cannot be represented as an array of datetimes, say because of an unparseable value or a mixture of timezones, the column or index will be returned unaltered as an object data type. For non-standard datetime parsing, use pd.to_datetime after pd.read_csv. To parse an index or column with a mixture of timezones, specify date_parser to be a partially-applied pandas.to_datetime() with utc=True. See Parsing a CSV with mixed timezones for more.

```python
# Analyser de la colonne à l'index 0 en tant que Datetime.
sales = pd.read_csv('RetailSales_BeerWineLiquor.csv',parse_dates=[0])
```

## Ré-échantillonnage

Une opération courante avec les données de séries temporelles est le ré-échantillonnage basé sur l'indice de la série temporelle. Voyons comment utiliser la méthode resample(). [[doc](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html)]

```python
# Notre index
sales.index
```

```python
# Ré-initialiser DATE à l'index
```

```python
sales = sales.set_index("DATE")
```

```python
sales
```

Lorsque vous appelez `.resample()`, vous devez d'abord passer un paramètre **rule**, puis vous devez appeler une sorte de fonction d'agrégation.

Le paramètre **rule** décrit la fréquence à laquelle appliquer la fonction d'agrégation (quotidienne, mensuelle, annuelle, etc.).
Il est passé en utilisant un "alias de décalage (ou offset)" - se référer au tableau ci-dessous. [[doc](http://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases)]

La fonction d'agrégation est nécessaire parce que, en raison du rééchantillonnage, nous avons besoin d'une règle mathématique pour joindre les lignes (moyenne, somme, nombre, etc.)

```python
# Moyennes annuelles
sales.resample(rule='A').mean()
```

La règle de ré-échantillonnage "A" prend tous les points de données d'une année donnée, applique la fonction d'agrégation (dans ce cas, nous calculons la moyenne) et rapporte le résultat comme étant le dernier jour de cette année. La note 2020 dans cet ensemble de données n'était pas complète.
