# Filtrage Conditionnel

## Importations

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

## Conditions

```python
# df['total_bill'] > 30
```

```python
bool_series = df['total_bill'] > 30
```

```python
df[bool_series]
```

```python
df[df['total_bill']>30]
```

```python
df[df['sex'] == 'Male']
```

## Conditions multiples

Rappelez-vous les étapes :

* Obtenir les conditions
* Encadrer chaque condition par des parenthèses
* Utilisez l'opérateur | ou &, selon que vous voulez un résultat de type 
    * OR | (l'une ou l'autre des conditions est vraie)
    * AND & (les deux conditions doivent être vraies)
* Vous pouvez également utiliser l'opérateur ~ comme une opération NOT.

```python
df[(df['total_bill'] > 30) & (df['sex']=='Male')]
```

```python
df[(df['total_bill'] > 30) & ~(df['sex']=='Male')]
```

```python
df[(df['total_bill'] > 30) & (df['sex']!='Male')]
```

```python
# Le Weekend
df[(df['day'] =='Sun') | (df['day']=='Sat')]
```

## Opérateur conditionnel isin()

Nous pouvons utiliser l'opérateur .isin() pour filtrer par une liste d'options.

```python
options = ['Sat','Sun']
df['day'].isin(options)
```

```python
df[df['day'].isin(['Sat','Sun'])]
```
